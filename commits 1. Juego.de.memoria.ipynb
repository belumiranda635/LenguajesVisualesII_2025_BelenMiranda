## Juego de memoria

# Commits 1. Breve explicacion.

1) ¿Que hace el programa?
Primeramente, el programa hace que el jugador debe voltear cartas, encontrar parejas antes de que termine el tiempo y ver su puntajes (intentos y aciertos).

2) El programa contiene:

- La principal libreria Tkinter(para crear ventanas, botones y etiquetas(UI)) lo que estamos utilizando para el proyecto y otras librerias como random (para mezclar las cartas aleatoriamente) y messagebox (para mostrar mensajes).
- Variables Globales
- Funciones como iniciar_juego(), voltear_carta(indice), comprobar_cartas(), actualizar_tiempo(), actualizar_contadores(), reiniciar()
- Temporizador (cuenta 60 segundos)
- Puntajes (intentos y aciertos)

```python
import tkinter as tk
import random
from tkinter import messagebox

# INTERFAZ
ventana = tk.Tk()
ventana.title("Juego de Memoria")

# VARIABLES GLOBALES
cartas = []
botones = []
primer_carta = None
segunda_carta = None
intentos = 0
aciertos = 0
tiempo = 60  
timer_activo = True

def iniciar_juego():
    global cartas, botones, primer_carta, segunda_carta, intentos, aciertos, tiempo, timer_activo
    
    primer_carta = None
    segunda_carta = None
    intentos = 0
    aciertos = 0
    tiempo = 60
    timer_activo = True
    
    cartas = list(range(1, 9)) * 2
    random.shuffle(cartas)
    
    for widget in frame_cartas.winfo_children():
        widget.destroy()

    botones.clear()

    for i in range(16):
        boton = tk.Button(frame_cartas, text="?", width=8, height=4,
                          command=lambda i=i: voltear_carta(i))
        boton.grid(row=i//4, column=i%4, padx=5, pady=5)
        botones.append(boton)
    
    actualizar_tiempo()
    actualizar_contadores()

def voltear_carta(indice):
    global primer_carta, segunda_carta
    
    boton = botones[indice]

    if boton['text'] != "?" or segunda_carta is not None:
        return
    
    boton['text'] = cartas[indice]
    
    if primer_carta is None:
        primer_carta = indice
    else:
        segunda_carta = indice
        ventana.after(600, comprobar_cartas)

def comprobar_cartas():
    global primer_carta, segunda_carta, intentos, aciertos
    
    if cartas[primer_carta] == cartas[segunda_carta]:
        botones[primer_carta]['state'] = 'disabled'
        botones[segunda_carta]['state'] = 'disabled'
        aciertos += 1
    else:
        botones[primer_carta]['text'] = "?"
        botones[segunda_carta]['text'] = "?"
    
    intentos += 1
    actualizar_contadores()

    primer_carta = None
    segunda_carta = None
    
    if aciertos == 8:
        messagebox.showinfo("¡Ganaste!", f"Completaste el juego!\nIntentos: {intentos}\nAciertos: {aciertos}")
        reiniciar()

def actualizar_contadores():
    label_intentos.config(text=f"Intentos: {intentos}")
    label_aciertos.config(text=f"Aciertos: {aciertos}/8")

def actualizar_tiempo():
    global tiempo, timer_activo
    if not timer_activo:
        return

    label_tiempo.config(text=f"Tiempo: {tiempo}s")

    if tiempo <= 0:
        messagebox.showwarning("Tiempo agotado", "¡Se acabó el tiempo!")
        reiniciar()
        return

    tiempo -= 1
    ventana.after(1000, actualizar_tiempo)

def reiniciar():
    iniciar_juego()

# Label
label_tiempo = tk.Label(ventana, text="Tiempo: 60s", font=("Arial", 14))
label_tiempo.pack(pady=5)

label_intentos = tk.Label(ventana, text="Intentos: 0", font=("Arial", 14))
label_intentos.pack()

label_aciertos = tk.Label(ventana, text="Aciertos: 0/8", font=("Arial", 14))
label_aciertos.pack(pady=5)

frame_cartas = tk.Frame(ventana)
frame_cartas.pack()
# Button
boton_reiniciar = tk.Button(ventana, text="Reiniciar Juego", font=("Arial", 12), command=reiniciar)
boton_reiniciar.pack(pady=10)

iniciar_juego()
ventana.mainloop()
```
